```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# for i in range(10):
# Path to your saved episode
episode_path = f"/homes/53/fpinto/jafar/data/coinrun_episodes/episode_23.npz"

# Load episode
data = np.load(episode_path)
frames = data["rgb"]          # shape: (T, H, W, C)
rewards = data["reward"]        # shape: (T,)
dones = data["done"]            # shape: (T,)
actions = data["actions"]          # shape: (T,)

if len(frames) > 0:
    # Print metadata
    print(f"Number of frames: {len(frames)}")
    print(f"Rewards: {rewards}")
    print(f"Dones: {dones}")
    print(f"Actions: {actions}")

    # Select every 20th frame and always include the final-1 frame
    indices = list(range(0, len(frames), 5))
    if indices[-1] != len(frames) - 1:
        indices.append(len(frames) - 2)

    selected_frames = frames[indices]
    selected_rewards = rewards[indices]
    selected_dones = dones[indices]

    # Plot each frame in a vertical column
    n = len(selected_frames)
    fig, axs = plt.subplots(n, 1, figsize=(3, n * 2.5))

    for i in range(n):
        axs[i].imshow(selected_frames[i])
        axs[i].axis("off")
        axs[i].set_title(f"r={selected_rewards[i]} d={int(selected_dones[i])}", fontsize=8)

    plt.tight_layout()
    plt.show()
    # break
```
